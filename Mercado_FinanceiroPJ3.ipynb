```python
# Libs Necessária

# Libs para Modelagem e Matrizes

import pandas as pd
import numpy as np

# Libs para análises Gráficas

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

# Lib para ignorar avisos

import warnings

#  Desabilitando avisos

warnings.filterwarnings('ignore')
```

```python
# Leitura dos Dados - head são os 5 primeiros

Base_Dados = pd.read_excel('/content/Vase_004 - Magalu - Sem Resolução.xlsx')

Base_Dados.head()
```

```python
# Dimensão

Base_Dados.shape
```

```python
# Verificar informação doa tipos de dados

Base_Dados.info()
```

```python
# Analisar as estatísticas

Base_Dados.describe()
```

```python
# Núlos

Base_Dados.isnull().sum()
```

```python
# Colocar a data no lugar do Index

Dados = Base_Dados.set_index('Data')

Dados.head()
```

```python
# Analíses Estatísticas - Séries Temporais

plt.style.use ('seaborn-darkgrid')
plt.figure(figsize=(16,5))
plt.title('Analise das Ações da Magalu - Fechamento', fontsize=15, loc='left')
plt.plot(Dados.index, Dados ['Fechamento'])

plt.xlabel('Período da Cotação')
plt.ylabel('Valor da Ação ($)');
```

```python
# Leitura dos Dados - tail são os 5 últimos

Base_Dados.tail()
```

```python
# Medir a média de 5 em 5 dias (ou quantos dias vc prentender até 30 dias)

Media_Movel = Dados['Fechamento'].rolling(5).mean()
Media_Tendencia = Dados['Fechamento'].rolling(30).mean()


plt.style.use ('seaborn-darkgrid')
plt.figure(figsize=(16,5))
plt.title('Analise das Ações da Magalu - Fechamento', fontsize=15, loc='left')

plt.plot(Dados.index, Dados ['Fechamento'])
plt.plot(Dados.index, Media_Movel)
plt.plot(Dados.index, Media_Tendencia)

plt.xlabel('Período da Cotação')
plt.ylabel('Valor da Ação ($)');
```

```python
sns.boxplot( data=Dados, x='Fechamento');
```

```python
# Boxplot Mensal

Base_Dados['Mes'] = Base_Dados ['Data'].dt.month

plt.figure(figsize=(16, 5))
sns.boxplot(data=Base_Dados, x='Mes', y='Fechamento');
```

```python
Base_Dados.groupby(['Mes']).describe()['Fechamento']
```

```python
Grafico = go.Figure(
    data=[
        go.Candlestick(
            x=Dados.index,
            open=Dados['Abertura'],
            high=Dados['Maior'],
            low=Dados['Menor'],
            close=Dados['Fechamento'],
        )
    ]
)

Grafico.update_layout(xaxis_rangeslider_visible=False)

Grafico.show()
```
