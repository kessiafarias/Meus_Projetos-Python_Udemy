```python
# Bibliotecas

# Labels de Modelagem e Matrizes
import pandas as pd
import numpy as np

# Labels de Análises Gráficas
import matplotlib.pyplot as plt
import seaborn as sns

# Labels de desativar avisos
import warnings
warnings.filterwarnings('ignore')
```

```python
# Lendo Base de Dados
Base_Dados = pd.read_excel('/content/Dados_Empresas_Energia.xlsx')

# Verificando
Base_Dados.head()
```

```python
# Séries Temporais

# Setar o index
Base_Dados.set_index('Data', inplace=True)

# Verificar
Base_Dados.head()
```

```python
# Gráfico

# Estilo
plt.style.use('seaborn-darkgrid')

# Tamanho
plt.figure(figsize=(16, 7))

# Título
plt.title('Análsie Ação de Empresas de Energia', loc='left', fontsize= 18, fontweight=0)

# Plot Petrobras
plt.plot(Base_Dados.index, Base_Dados['Petrobras'], color='#008c4a', linewidth=4, alpha=0.7 )

# Texto da Petrobras
plt.text(Base_Dados.index[-1], Base_Dados['Petrobras'].tail(1), 'Petrobras', color='#008c4a', size= 'large', horizontalalignment = 'left')

# Plot de Todas as colunas
for Coluna in Base_Dados.columns[1:]:

  # Plot das outras ações
  plt.plot(Base_Dados.index, Base_Dados[Coluna], color='gray', linewidth=2, alpha=0.7 )

  # Texto das outras ações
  plt.text(Base_Dados.index[-1], Base_Dados[Coluna].tail(1), Coluna, color='gray', size= 'large', horizontalalignment = 'left')

# Ajuse dos Labels laterais
plt.xlabel('Período')
plt.ylabel('Preço do Fechamento (R$)');
```
