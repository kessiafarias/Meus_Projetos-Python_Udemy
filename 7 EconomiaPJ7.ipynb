```python
# Importando Bibliotecas

# Labels para Modelagem e Matrizes

import pandas as pd
import numpy as np

# Labels para Análises Gráficas

import matplotlib.pyplot as plt
import seaborn as sns

# Label

import warnings


# Desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
# Ler Dados

Base_Dados = pd.read_excel('/content/Dados_Pib.xlsx')

Base_Dados.head(10)
```

```python
Base_Dados.groupby(by= ['Territorialidades', 'Ano']).mean()
```

```python
# Sitema de Grids

#Criar Sistema de Grids

Grid_Graficos = sns.FacetGrid(data=Base_Dados, col='Territorialidades', hue='Territorialidades', col_wrap=4)

# Cor de Fundo

Cor_Fundo = '#f5f5f5'

#Adicionar Gráfico de linhas em cada Gráfico

Grid_Graficos = Grid_Graficos.map(plt.plot, 'Ano', 'PIB per capita')

# Adicionar Sombra + o ajuste do título

Grid_Graficos = Grid_Graficos.map(plt.fill_between, 'Ano', 'PIB per capita', alpha=0.2).set_titles('{col_name} Territorialidades')

# Filtrar o título

Grid_Graficos = Grid_Graficos.set_titles('{col_name}')

# Adicionar subtitulo

Grid_Graficos = Grid_Graficos.fig.suptitle('Python para Análise de Dados \n Evolução de Renda Per Capta por Estado', fontsize=18)

# Ajustando título

plt.subplots_adjust(top=0.92)

# Rodapé

Rodape = '''
Esse relatório foi elaborado no treinamento "Python para análise de Dados"
by = Maria Késsia
'''

# Incluíndo o Rodapé no relatório

plt.figtext(0.5, -0.05, Rodape, ha='center', va='bottom', size=12, color='#938ca1');
```
