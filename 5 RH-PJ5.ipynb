```python
# Labels Necessárias

# Label para Modelages e Matrizes

import pandas as pd
import numpy as np


# Labels para Análises Gráficas

import matplotlib.pyplot as plt
import seaborn as sns


# Label para Ignorar

import warnings

# Desabilitando Avisos

warnings.filterwarnings('ignore')
```

```python
# Lendo os Dados

Base_Dados = pd.read_csv('/content/Salary_Data.csv')


# Verificando

Base_Dados.head()
```

```python
# Renomeando as Colunas

Base_Dados.rename(columns={
    'YearsExperience' : 'XP',
    'Salary' : 'Renda',
}, inplace=True)
```

```python
Base_Dados.shape
```

```python
 # Campos Núlos

 Base_Dados.isnull().sum()
```

```python
# Gráfico de Núlos

sns.heatmap(Base_Dados.isnull(), cbar=False)
```

```python
# Estatística

Base_Dados.describe()
```

```python
#  Plotagem

plt.figure(figsize=(10,5))
plt.title('Análise de Renda')
sns.kdeplot(Base_Dados['Renda'], color='green');
```

```python
plt.figure(figsize=(10,5))
plt.title('Análise de XP')
sns.kdeplot(Base_Dados['XP'], color='Blue');
```

```python
# Distribuição

plt.figure(figsize=(10,5))
sns.boxplot(Base_Dados['Renda'] );
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(Base_Dados['XP'] );
```

```python
#Distribuição

plt.figure(figsize=(10,5))
sns.scatterplot(data=Base_Dados, x='Renda', y='XP');
```

```python
plt.figure(figsize=(10,5))
sns.regplot(data=Base_Dados, x='Renda', y='XP');
```

```python
# Correlação pearson

Correlacao = Base_Dados.corr('pearson')
plt.figure(figsize=(10,5))
sns.heatmap(Correlacao, annot=True, cbar=True)
```

```python
# Correlação spearman

Correlacao = Base_Dados.corr('spearman')
plt.figure(figsize=(10,5))
sns.heatmap(Correlacao, annot=True, cbar=True);
```
